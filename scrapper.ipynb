```python
#installed library dputils
from dputils.scrape import Scraper, Tag
#scraper website se data collect krta h
#tag desired details tag krta h
```

```python
url= 'https://www.flipkart.com/search?q=bags&page=10'  #&page=''se page no. ka data hoga
sc= Scraper(url)  #flipkart ka page sc me pass kra url k through
```

```python
print(sc.soup)    #soup= website wala data 
```

```python
target = Tag('div', cls='_1YokD2 _3Mn1Gg') #important
items= Tag('div', cls= '_1xHGtK _373qXS') #important repeating item
brand= Tag('div', cls='_2WkVRV')
details= Tag('a',cls='IRpwTa')
price= Tag('div', cls='_30jeq3')
color= Tag('div', cls='_3eWWd-')
discount= Tag('div', cls='_3Ay6Sb')
```

```python
data=sc.get_all(target,items, brand=brand , detail=details, price=price, color=color, discount=discount)  #pehla wala naam hai, dusra wala code hai
print(data)
print(len(data))
```
